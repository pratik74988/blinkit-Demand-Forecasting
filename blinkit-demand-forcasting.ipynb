```python


merge_ords_ordID = pd.merge(orders, order_items, on='order_id', how='outer', indicator=True)

print(" ")
merge_ords_ordID['_merge'].value_counts()
```

```python
merge_ords_ordID.head()
```

```python
daily_orders_on_product = merge_ords_ordID[['order_id', 'order_date', 'payment_method', 'product_id', 'quantity']]
daily_orders_on_product
```

```python
daily_orders_on_product['order_date'].dtype
```

```python
products_daily = daily_orders_on_product[['order_date', 'product_id','quantity']]
products_daily
```

```python
products_daily['product_id'].unique().size
```

```python
merge_ords_ordID['order_date'] = pd.to_datetime(merge_ords_ordID['order_date'])
```

```python
merge_ords_ordID['order_date'] = pd.to_datetime(
    merge_ords_ordID['order_date']
).dt.normalize()
```

```python
grouped = merge_ords_ordID.groupby(['order_date', 'product_id'])

daily_product_sales = grouped['quantity'].sum().reset_index()
daily_product_sales
```

```python
daily_product_sales['order_date'].unique().size
```

```python
daily_product_sales['order_date'].unique()
```

```python
full_dates = pd.date_range(daily_product_sales['order_date'].min(), daily_product_sales['order_date'].max())
full_dates
```

```python
products = daily_product_sales['product_id'].unique()

full_index = daily_product_sales['product_id'].unique()

full_index = pd.MultiIndex.from_product(
    [products,full_dates],
    names=['product_id','date']
)
```

```python
full_index
```

```python
full_df = pd.DataFrame(index=full_index).reset_index()
```

```python
full_df.rename(columns={'date': 'order_date'}, inplace=True)
```

```python
full_df
```

```python
df_full = full_df.merge(daily_product_sales,
                       on =['product_id','order_date'],
                       how='left'
                       )
df_full
```

```python
df_full[df_full['quantity'].isna()]
```

```python
df_full['quantity'] = df_full['quantity'].fillna(0)

df_full.isna().sum()
```

```python
df_full = df_full.rename(columns={'order_date':'date'})
```

```python
df_full.head()
```

```python
df_full = df_full.sort_values(['product_id','date'])
```
