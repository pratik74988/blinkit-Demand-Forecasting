```python


merge_ords_ordID = pd.merge(orders, order_items, on='order_id', how='outer', indicator=True)

print(" ")
merge_ords_ordID['_merge'].value_counts()
```

```python
merge_ords_ordID.head()
```

```python
daily_orders_on_product = merge_ords_ordID[['order_id', 'order_date', 'payment_method', 'product_id', 'quantity']]
daily_orders_on_product
```

```python
daily_orders_on_product['order_date'].dtype
```

```python
products_daily = daily_orders_on_product[['order_date', 'product_id','quantity']]
products_daily
```

```python
products_daily['product_id'].unique().size
```

```python
merge_ords_ordID['order_date'] = pd.to_datetime(merge_ords_ordID['order_date'])
```

```python
merge_ords_ordID['order_date'] = pd.to_datetime(
    merge_ords_ordID['order_date']
).dt.normalize()
```

```python
grouped = merge_ords_ordID.groupby(['order_date', 'product_id'])

daily_product_sales = grouped['quantity'].sum().reset_index()
daily_product_sales
```
